## Show Image Files

```python
import celestializer as cl

img = cl.RawImage(cl.Paths.observations / "2024-11-05/IMG_3320.CR2").load()
```

### Raw Image Data

Use `preview()` to downscale the image and `show()` to display the image.

```python
print("Image size", (img.metadata.width, img.metadata.height))
img.raw().preview().show()
```

### Comparison of RGB conversion methods

```python
img.to_rgb_rawkit().preview().show()
img.to_rgb_cv2().preview().show()
```

### Mask colors in raw image

Assumes the filter pattern of the Bayer filter is RGGB.

<!--  -->

```python
img_green = img.raw().channel("green")
img_green.preview().show()
img_green.save(cl.Paths.saved / "sky_green_only.png")
```

```python
img_green = img.raw().green_interpolated()
img_green.save(cl.Paths.saved / "sky_green_interpolated.png")
img_green.preview().show()
```
